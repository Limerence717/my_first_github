### homework5：使用数组的索引和切片，掌握数组数值修改

一维数组的操作方法

```python
a = np.arange(10)
print(a)
# 冒号分隔切片参数 start:stop:step 来进行切片操作print(a[2:7:2])# 从索引 2 开始到索引 7 停止，间隔为2
print(a[2:7:2])
# 如果只放置一个参数，如 [2]，将返回与该索引相对应的单个元素
print(a[2])
# 如果为 [2:]，表示从该索引开始以后的所有项都将被提取
print(a[2:])
```

多维数组的操作方法

```python
# 创建数组
t1 = np.arange(24).reshape(4, 6)
print("原始数组 t1:")
print(t1)

# 取一行
print("\n取第 2 行:")
print(t1[1])
print(t1[1, :])  # 取第2行，:表示取该行的所有列

# 取连续的多行
print("\n从第 2 行开始取到最后一行:")
print(t1[1:])
print("\n取第 2 行到第 3 行:")
print(t1[1:3, :])

# 取不连续的多行
print("\n取第 1 行、第 3 行和第 4 行:")
print(t1[[0, 2, 3]])
print(t1[[0, 2, 3], :])

# 取一列
print("\n取第 2 列:")
print(t1[:, 1])

# 取连续的多列
print("\n从第 2 列开始取到最后一列:")
print(t1[:, 1:])

# 取不连续的多列
print("\n取第 1 列、第 3 列和第 4 列:")
print(t1[:, [0, 2, 3]])

# 取某一个值
print("\n取第 3 行第 4 列的值:")
print(t1[2, 3])
```

 数组中的数值修改

```python
# 创建数组
t = np.arange(24).reshape(4, 6)
print("原始数组 t:")
print(t)

# 修改某一行的值
t[1, :] = 0
print("\n修改第 2 行的值:")
print(t)

# 修改某一列的值
t = np.arange(24).reshape(4, 6)
t[:, 1] = 0
print("\n修改第 2 列的值:")
print(t)

# 修改连续多行
t = np.arange(24).reshape(4, 6)
t[1:3, :] = 0
print("\n修改第 2 行到第 3 行的值:")
print(t)

# 修改连续多列
t = np.arange(24).reshape(4, 6)
t[:, 1:4] = 0
print("\n修改第 2 列到第 4 列的值:")
print(t)

# 修改多行多列
t = np.arange(24).reshape(4, 6)
t[1:4, 2:5] = 0
print("\n修改第 2 行到第 4 行，第 3 列到第 5 列的值:")
print(t)

# 修改多个不相邻的点
t = np.arange(24).reshape(4, 6)
t[[0, 1], [2, 3]] = 0
print("\n修改多个不相邻的点:")
print(t)

# 根据条件修改
t = np.arange(24).reshape(4, 6)
t[t < 10] = 0
print("\n将小于 10 的值设置为 0:")
print(t)

# 使用逻辑判断
# 逻辑与
t = np.arange(24).reshape(4, 6)
t[(t > 2) & (t < 6)] = 0
print("\n逻辑与 (t > 2 且 t < 6) 修改后的数组:")
print(t)

# 逻辑或
t = np.arange(24).reshape(4, 6)
t[(t < 2) | (t > 6)] = 0
print("\n逻辑或 (t < 2 或 t > 6) 修改后的数组:")
print(t)

# 逻辑非
t = np.arange(24).reshape(4, 6)
t[~(t > 6)] = 0
print("\n逻辑非 (~(t > 6)) 修改后的数组:")
print(t)

# 使用 clip 函数
t = np.arange(24).reshape(4, 6)
t = t.clip(10, 18)
print("\n使用 clip 函数限制值在 [10, 18] 范围内:")
print(t)
```

```python
# 三目运算（ np.where(condition, x, y)满足条件(condition)，输出x，不满足输出y。)）
score = np.array([[80, 88], [82, 81], [75, 81]])
result = np.where(score > 80, True, False)
print(result)
```

### homework6：使用数组添加，删除，去重

数组的添加

```python
# 1. numpy.append 函数在数组的末尾添加值。追加操作会分配整个数组，并把原来的数组复制到新数组中。
# 此外，输入数组的维度必须匹配否则将生成ValueError。
'''
参数说明：
arr：输入数组
values：要向arr添加的值，需要和arr形状相同（除了要添加的轴）
axis：默认为 None。当axis无定义时，是横向加成，返回总是为一维数组！
当axis有定义的时候，分别为0和1的时候（列数要相同）。当axis为1时，数组是加在右边（行数要相同）。
'''
# 创建数组
a = np.array([[1, 2, 3], [4, 5, 6]])
print('第一个数组：')
print(a)

# 默认情况（axis=None）
print('\n向数组添加元素：')
print(np.append(a, [7, 8, 9]))

# 沿轴 0 添加元素
print('\n沿轴 0 添加元素：')
print(np.append(a, [[7, 8, 9]], axis=0))

# 沿轴 1 添加元素
print('\n沿轴 1 添加元素：')
print(np.append(a, [[5, 5, 5], [7, 8, 9]], axis=1))
```

```python
# 2. numpy.insert
# 函数在给定索引之前，沿给定轴在输入数组中插入值。
# 如果值的类型转换为要插入，则它与输入数组不同。
# 插入没有原地的，函数会返回一个新数组。 此外，如果未提供轴，则输入数组会被展开。
# 创建数组
a = np.array([[1, 2], [3, 4], [5, 6]])
print('第一个数组：')
print(a)

# 未传递 axis 参数
print('\n未传递 Axis 参数。 在插入之前输入数组会被展开。')
print(np.insert(a, 3, [11, 12]))

# 传递 axis 参数
print('\n传递了 Axis 参数。 会广播值数组来配输入数组。')
print('沿轴 0 广播：')
print(np.insert(a, 1, [11], axis=0))

print('\n沿轴 1 广播：')
print(np.insert(a, 1, 11, axis=1))
```

数组中的删除

```python
# numpy.delete 函数返回从输入数组中删除指定子数组的新数组。 与 insert() 函数的情况一样，
# 如果未提供轴参数，则输入数组将展开。
'''
参数说明：arr： 输 入 数 组
obj：可以被切片，整数或者整数数组，表明要从输入数组删除的子数组axis：沿着它删除给定子数组的轴，如果未提供，则输入数组会被展开
'''
# 创建数组
a = np.arange(12).reshape(3, 4)
print('第一个数组：')
print(a)

# 未传递 axis 参数
print('\n未传递 Axis 参数。 在删除之前输入数组会被展开。')
print(np.delete(a, 5))
print('删除多个元素：')
print(np.delete(a, [1, 3, 5], axis=None))

# 沿轴 0 删除元素
print('删除第二行：')
print(np.delete(a, 1, axis=0))

# 沿轴 1 删除元素
print('\n删除每一行中的第二列：')
print(np.delete(a, 1, axis=1))
```

数组去重

```python
# numpy.unique 函数用于去除数组中的重复元素。
'''
arr：输入数组，如果不是一维数组则会展开
return_index：如果为 true，返回新列表元素在旧列表中的位置（下标），并以列表形式储存
return_inverse：如果为 true，返回旧列表元素在新列表中的位置（下标），并以列表形式储存
return_counts：如果为 true，返回去重数组中的元素在原数组中的出现次数
'''
# 创建数组
a = np.array([5, 2, 6, 2, 7, 5, 6, 8, 2, 9])
print('第一个数组：')
print(a)

# 去重值
print('\n第一个数组的去重值：')
u = np.unique(a)
print(u)

# 去重值的索引
print('\n去重数组的索引数组：')
u, indices = np.unique(a, return_index=True)
print(indices)

# 去重值的逆索引
print('\n我们可以看到每个和原数组下标对应的数值：')
print(a)
print('去重数组的下标：')
u, indices = np.unique(a, return_inverse=True)
print(u)
print(indices)

# 去重值的重复数量
print('\n返回去重元素的重复数量：')
u, indices = np.unique(a, return_counts=True)
print(u)
print(indices)
```

### homework7：使用最大，最小，平均值等各种计算

```python
# 创建数组
score = np.array([[80, 88], [82, 81], [75, 81]])
print(score)

# 获取所有数据的最大值
result = np.max(score)
print("所有数据的最大值:", result)

# 获取某一个轴上的数据最大值
result = np.max(score, axis=0)
print("每一列的最大值:", result)

# 获取所有数据的最小值
result = np.min(score)
print("所有数据的最小值:", result)

# 获取某一个轴上的数据最小值
result = np.min(score, axis=0)
print("每一列的最小值:", result)

# 数据的比较
# 第一个参数中的每一个数与第二个参数比较返回大的
result = np.maximum([-2, -1, 0, 1, 2], 0)
print("np.maximum([-2, -1, 0, 1, 2], 0):", result)

# 第一个参数中的每一个数与第二个参数比较返回小的
result = np.minimum([-2, -1, 0, 1, 2], 0)
print("np.minimum([-2, -1, 0, 1, 2], 0):", result)

# 接受的两个参数，也可以大小一致；第二个参数只是一个单独的值时，其实是用到了维度的广播机制
result = np.maximum([-2, -1, 0, 1, 2], [1, 2, 3, 4, 5])
print("np.maximum([-2, -1, 0, 1, 2], [1, 2, 3, 4, 5]):", result)

# 获取所有数据的平均值
result = np.mean(score)
print("所有数据的平均值:", result)

# 计算每一列的平均值
result = np.mean(score, axis=0)
print("每一列的平均值:", result)

# 计算每一行的平均值
result = np.mean(score, axis=1)
print("每一行的平均值:", result)
```
